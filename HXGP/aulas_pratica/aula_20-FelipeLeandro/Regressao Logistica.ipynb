```
from sklearn.linear_model import  LogisticRegression
import numpy as np
```

```
# Dados de treino
# Conjunto de dados que representam as notas de cada aluno
x = np.array([[18,17,15],[2,2,2],[2,2,2],[4,2,6]])

# Representa os targets​
y = np.array([0,0,1])

###Rotaciona os Dados (Transposta dos dados de treino)
x = np.rot90(x)
```

```
# Construindo o modelo
model = LogisticRegression()

# Aplicando o treinamento
model.fit(x, y)

# Predição dos Valores
y_pred = model.predict(x)

# Métrica de Acurácia
print("Acuracia:", model.score(x, y))
```
