```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import perceptron
```

```
# Dados de treino
# Conjunto de dados que representam as notas de cada aluno
x = np.array([[18,17,15],[2,2,2],[2,2,2],[4,2,6]])

# Representa os targets​
y = np.array([0,0,1])
```

```
###Rotaciona os Dados (Transposta dos dados de treino)
x = np.rot90(x)
##Criando o modelo de predição​
## max_iter : Quantidade máxima de épocas as serem utilizadas​
## eta : taxa de aprendizagem
net = perceptron.Perceptron(max_iter=100,eta0=0.1)
net.fit(x, y)
```

```
print ("Saída Esperada  " + str(net.predict(x)))
print ("Saída Atual     " + str(y))
print ("Acuracia        " + str(net.score(x, y) * 100) + "%")
print('Pesos: ' + str(net.coef_))
```
