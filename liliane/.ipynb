```python
columns_to_read = ["id",
                   "data_inversa",
                   "horario",
                   "uf",                                
                   "br",              
                    "km",                                
                    "municipio",                         
                    "causa_acidente",                    
                    "tipo_acidente",                     
                    "classificacao_acidente",            
                    "fase_dia",                          
                    "sentido_via",                       
                    "condicao_metereologica",            
                    "tipo_pista",                        
                    "tracado_via",                       
                    "uso_solo",                                            
                    "pessoas",                            
                    "mortos",                             
                    "feridos_leves",                      
                    "feridos_graves",                     
                    "ilesos",                             
                    "ignorados",                          
                    "feridos",                            
                    "veiculos"] 
```

```python
import glob

all_files = glob.glob("acidentes-br/" + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, sep = ';', decimal=",", encoding = "ISO-8859-1", usecols=columns_to_read)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
data = pd.read_csv('acidentes-rodovia-br.csv')
data.drop(['Unnamed: 0'], axis=1, inplace=True)
data.head(2)
```

```python
data.id = data.id.astype('int32')
data.pessoas = data.pessoas.astype('uint8')
data.mortos = data.mortos.astype('uint8')
data.feridos_leves = data.feridos_leves.astype('uint8')
data.feridos_graves = data.feridos_graves.astype('uint8')
data.ilesos = data.ilesos.astype('uint8')
data.ignorados = data.ignorados.astype('uint8')
data.feridos = data.feridos.astype('uint8')
data.veiculos = data.veiculos.astype('uint8')
data.uf = data.uf.astype('category')
data.br = data.br.astype('category')
data.municipio = data.municipio.astype('category')
data.causa_acidente = data.causa_acidente.astype('category')
data.tipo_acidente = data.tipo_acidente.astype('category')
data.data = data.classificacao_acidente.astype('category')
data.fase_dia = data.fase_dia.astype('category')
data.sentido_via = data.sentido_via.astype('category')
data.condicao_metereologica = data.condicao_metereologica.astype('category')
data.tipo_pista = data.tipo_pista.astype('category')
data.tracado_via = data.tracado_via.astype('category')
data.uso_solo = data.uso_solo.astype('category')
```

```python
data['data_hora'] = data['data_inversa'].map(str) + ' ' + data['horario']
data['data_hora'] = pd.to_datetime(data['data_hora'])
data['ano'] = data['data_hora'].dt.year
data['mes'] = data['data_hora'].dt.month
data['hora'] = data['data_hora'].dt.hour
data.drop(['data_inversa', 'horario'], axis=1, inplace = True)
```
