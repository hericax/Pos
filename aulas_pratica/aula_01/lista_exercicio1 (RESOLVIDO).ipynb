EXERCÍCIO 1 - FAÇA UM SCRIPT VERIFIQUE SE UM NÚMERO (*num*) É PAR OU IMPAR:

```
import random
num = random.randint(1, 100)

# SE A DIVISÃO POR 2 FOR IGUAL A 0...
if (num % 2 == 0):
  print(num, " é par!")
else:
  print(num, "é impar!")
```

EXERCÍCIO 2 - FAÇA UMA SCRIPT QUE CALCULA UMA EXPRESSÃO MATEMÁTICA DADO 2 INTEIROS (*num1, num2 e op*) E UM OPERADOR (String = '+', '-', '*', '/'):

```
import random
num1 = random.randint(0, 100)
num2 = random.randint(0, 100)
op = ['+', '-', '*', '/'][random.randint(0, 3)]

# INICIALIZA UM OBJETO PARA ARMAZENAR O RESULTADO
resultado = 0

# VERIFICA CADA OPERADOR E REALIZA O CALCULA DE ACORDO COM O MESMO
if op == "+":
  resultado = num1 + num2
elif op == "-":
  resultado = num1 - num2
elif op == "*":
  resultado = num1 * num2
elif op == "/":
  # VALIDA DIVISÃO POR 0
  if num2 == 0:
    resultado = "ERRO: Divisão por 0!"
  else:
    resultado = num1 / num2

print(num1, op, num2, "=", resultado)
```

EXERCÍCIO 3 - FAÇA UM SCRIPT QUE IDENTIFIQUE SE UM NÚMERO INTEIRO (*num*) É PRIMO OU NÃO

```
import random
num = random.randint(1, 100)

# OBJETO QUE ARMAZENA A QUANTIDADE DE divisores
divisores = 0

# OBJETO divisor VARIA DE 1 ATE num
for divisor in range(1, num):

  # SE A DIVISÃO RETORNAR 0, OU SEJA, É DIVISÍVEL
  if num % divisor == 0:
    # INCREMENTA divisores
    divisores = divisores + 1
    # SE O NÚMERO DE DIVISORES FOR MAIOR QUE 1, OU SEJA, NÃO É PRIMO MAIS
    if divisores > 1:
      # ENCERRA O LAÇO for
      break

# VERIFICA A QUANTIDADE DE divisores
if divisores > 1:
  print(num, "não é primo!")
else:
  print(num, "é primo!")
```

EXERCÍCIO 4 - FAÇA UM SCRIPT QUER RETORNE DE UMA LISTA DE INTEIROS (*lista*):

*   MAIOR ELEMENTO
*   MENOR ELEMENTO
*   TAMANHO

**DESAFIO:** NÃO USE FUNÇÕES *min(), max() e len()*!

```
import random

lista = [random.randint(0,10) for i in range(random.randint(1,10))]

print(lista)

# DEFINE O OBJETO PARA ARMAZENAR O TAMANHO DA LISTA
tamanho = 1

# CONSIDERA O PRIMEIRO ELEMENTO DA LISTA COMO O MAIOR (INICIALIZAÇÃO)
maior = lista[0]

# CONSIDERA O PRIMEIRO ELEMENTO DA LISTA COMO O MENOR (INICIALIZAÇÃO)
menor = lista[0]

# PERCORRE A LISTA, DESCONSIDERANDO O PRIMEIRO ELEMENTO QUE JÁ FOI ARMAZENADO EM maior E menor
for e in lista[1:]:
  # VERIFICA O MAIOR
  if maior < e:
    maior = e
  # VERIFICA O MENOR
  if menor > e:
    menor = e
  # ATUALIZA O TAMANHO DA LISTA
  tamanho += 1

print("Maior: ", maior)
print("Menor: ", menor)
print("Tamanho: ", tamanho)
```

EXERCÍCIO 5 - FAÇA UM SCRIPT QUE RETORNE DE UMA LISTA DE INTEIROS (*lista*):

* A MEDIA
* A MEDIANA
* A MODA

```
import random

lista = [random.randint(0,10) for i in range(5)]

print("------------------------------------------------")
print("Lista origem: ", lista)

# OBJETOS PARA OS CALCULOS
soma = 0
tamanho = 0
maior_qtde = 0
qtde_elemento = {}

# ORDENA A LISTA (PARA A MEDIANA)
lista.sort()
print("Lista ordenada:", lista)
print("------------------------------------------------\n")

# LAÇO PARA VALORES IMPORTANTES (soma, tamanho e quantidade de elementos unicos)
for e in lista:  
  
  # SOMA E TAMANHO DA LISTA
  soma += e
  tamanho += 1
  
  # ARMAZENA A QUANTIDADE DE VEZES QUE UM ELEMENTO APARECE NA LISTA
  if e not in qtde_elemento:
    qtde_elemento[e] = 0
  qtde_elemento[e] += 1 
  
  # ARMAZENA A MAIOR QUANTIDADE DE VEZES QUE UM ELEMENTO POSSA APARECER NA LISTA
  if maior_qtde < qtde_elemento[e]:
    maior_qtde = qtde_elemento[e]

print("----------DADOS DA LISTAS-----------------------")
print("Soma: ", soma)
print("Tamanho: ", tamanho)
print("Qtde elementos: ", qtde_elemento)
print("Maior qtde: ", maior_qtde)
print("------------------------------------------------\n")

# MEDIA
media = soma / tamanho

# MEDIANA
# TAMANHO PAR
if tamanho % 2 == 0:
    # CALCULO DA MEDIANA CASO A LISTA TENHA TAMANHO PAR
    mediana = (lista[round(tamanho/2)] + lista[round(tamanho/2) + 1]) / 2 
# TAMANHO IMPAR
else:
  # CALCULO DA MEDIANA CASO A LISTA TENHA TAMANHO IMPAR
    mediana = lista[round(tamanho/2)]

# MODA 
modas = []
for e in qtde_elemento:
  if qtde_elemento[e] == maior_qtde:
    modas.append(e)

print("----------ESTATISTICA---------------------------")
print("Media: ", media)
print("Mediana: ", mediana)
print("Moda(s): ", modas)
print("------------------------------------------------\n")
```

EXERCÍCIO 6 - FAÇA UMA FUNÇÃO QUE TENHA COMO ENTRADA UMA LISTA DE INTEIROS (*idades*) E RETORNE UMA LISTA COM AS FAIXAS DE CADA IDADE, DE ACORDO COM AS REGRAS A SEGUIR:<br>
* '<18', 
* '18 A 21', 
* '21 A 27', 
* '28 A 35', 
* '36 A 50', 
* '51 A 65', 
* '65+'

```
import random

idades = [random.randint(5, 95) for i in range(100)]

faixas = []
for idade in idades:
  if idade < 18:
    faixas.append("<18")
  elif idade < 22:
    faixas.append("18 A 21")
  elif idade < 28:
    faixas.append("21 A 27")
  elif idade < 36:
    faixas.append("28 A 35")
  elif idade < 51:
    faixas.append("36 A 50")
  elif idade < 66:
    faixas.append("51 A 65")
  else:
    faixas.append("65+")

for i in range(len(idades)):
  print("Idade: ", idades[i], " - Faixa: ", faixas[i])
```

EXERCÍCIO 7 - DADA UM LISTA (*cabecalho*) CONTENDO O CABEÇALHO DE UMA TABELA, FAÇA AS SEGUINTES TRANSFORMAÇÕES:<br>
* Remova espaços a direita e a esquerda (função \<string\>.strip())
* Deixe todas a letras maiúsculas (função \<string\>.upper())
* substitua " " (espaços) por "_" (função \<string\>.replace())

```
cabecalho = ["nome ", "IdadE", "data de nascimento", "endereco", " cpf", "Valor Devido", "Valor ja pago"]

print("Antes: ", cabecalho)

# PERCORRE A LISTA CABECALHO, INDICE A INDICE (i)
for i in range(len(cabecalho)):
  # REALIZA AS MODIFICAÇÕES
  cabecalho[i] = cabecalho[i].strip().upper().replace(" ","_")

print("Depois: ", cabecalho)
```
